## 4. Basic Data Analysis

```python
df.groupby("Category")["Sales"].sum()
```

## 5. Bar Chart

```python
df.groupby("Category")["Sales"].sum().plot(kind="bar")
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
```

## 6. Histogram

```python
df["Sales"].plot(kind="hist")
plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

## 7. Box Plot

```python
df.boxplot(column="Sales")
plt.title("Box Plot of Sales")
plt.show()

```

## 8. Correlation Heatmap

```python
correlation=df.corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(correlation,annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

## 9. Sales Trend Over Time (Line Chart)

```python
df["Order Date"]=pd.to_datetime(df["Order Date"])
sales_by_date=df.groupby("Order Date")["Sales"].sum()
sales_by_date.plot(figsize=(12,5))
plt.title("Sales Trend Over Time")
plt.xlabel("Order Date")
plt.ylabel("Total Sales")
plt.show()
```

## 10. Key Insights

1. The Superstore dataset contains 9,994 rows and 21 columns, providing sufficient data for analysis.

2. The dataset contains no missing values and no duplicate records, indicating that the data is clean and ready for analysis.

3. Technology generated the highest total sales among all product categories.

4. The histogram shows that most sales are of lower values, while only a few orders have very high sales.

5. The box plot indicates the presence of outliers in the Sales column, meaning a small number of orders have exceptionally high sales.
